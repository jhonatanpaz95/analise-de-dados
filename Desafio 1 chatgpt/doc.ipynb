```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
loja1_df = pd.read_csv('loja1.txt')
```

```python
loja1_agrupado_df = loja1_df.drop(columns=['Mês'])
loja1_agrupado_df = loja1_agrupado_df.groupby('Produto', as_index=False).agg({
    'Preço Unitário': 'mean',
    'Quantidade Vendida': 'sum',
    'Faturamento Total': 'sum'
})

totais = loja1_agrupado_df.select_dtypes(include=[pd.np.number]).sum().rename('Total')
loja1_agrupado_df.append(totais)

sns.set(rc={'figure.figsize': (14,6)})
fig = sns.barplot(data=loja1_agrupado_df, x='Produto', y="Faturamento Total", hue="Produto", width=0.5, errorbar=None)
fig.set_title('LOJA 01', fontsize=10, loc='center')
for p in fig.patches:

    fig.annotate(format(p.get_height(), '.2f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9),
                textcoords = 'offset points',
                fontsize=9)
    
sns.lineplot(data= loja1_agrupado_df, x='Produto', y='Faturamento Total')
display(loja1_agrupado_df)
```

```python
vendas_janeiro_df = loja1_df.loc[loja1_df['Mês'] == 'Janeiro']
display(vendas_janeiro_df)
fig = sns.barplot(data=vendas_janeiro_df, x='Produto', y="Faturamento Total", hue="Produto", width=0.5, errorbar=None)

fig.set_title('Vendas Janeiro', fontsize=10, loc='center')

for c in fig.patches:
    fig.annotate(format(c.get_height(), '.2f'),
                 (c.get_x() + c.get_width() / 2., c.get_height()),
                 ha = 'center', va = 'center',
                 xytext = (0, 9),
                 textcoords = 'offset points')
```

TOTAL DE FATURAMENTO MENSAL

```python
faturamento_janeiro = vendas_janeiro_df['Faturamento Total'].sum()
faturamento_fevereiro = vendas_fevereiro_df['Faturamento Total'].sum()
faturamento_marco = vendas_marco_df['Faturamento Total'].sum()
#print(f'faturamento Janeiro = R$ {faturamento_janeiro:.2f}, \nfaturamento Fevereiro = R$ {faturamento_fevereiro:.2f}, \nfaruamento março = R$ {faturamento_marco:.2f}')

faturamento ={
    'janeiro' : faturamento_janeiro,
    'fevereiro' : faturamento_fevereiro,
    'março' : faturamento_marco
}

faturamento_df = pd.DataFrame(list(faturamento.items()), columns = ['Mês', 'Faturamento'])
display(faturamento_df)

fig= sns.barplot(data= faturamento_df, x='Mês', y='Faturamento', width=0.5, hue='Mês')
fig.set(title='Relação de faturamentos')

for c in fig.patches:
    fig.annotate(format(c.get_height(), '.2f'),
                 (c.get_x() + c.get_width() / 2., c.get_height()),
                 ha = 'center', va= 'center',
                 xytext = (0,9),
                 textcoords= 'offset points')
    
sns.lineplot(faturamento_df, linewidth=3,)
```

```python
faturamentos = [faturamento_janeiro, faturamento_fevereiro, faturamento_marco]

maior_faturamento = max(faturamentos)
print(maior_faturamento)
```

```python

janeiro_fevereiro = faturamento_janeiro - faturamento_fevereiro
janeiro_marco = faturamento_janeiro - faturamento_marco
fevereiro_marco = faturamento_fevereiro - faturamento_marco
print(janeiro_fevereiro, janeiro_marco, fevereiro_marco)


```

Saber a porcentagem de queda 

```python
porcentagem_de_queda = [(maior_faturamento - faturamento) / maior_faturamento * 100 for faturamento in faturamentos]

for i, porcentagem in enumerate(porcentagem_de_queda):
    print(f'Faturamento {i+1}: {faturamentos[i]} - Queda de {porcentagem:.2f}% em relação ao maior faturamento.')
```
