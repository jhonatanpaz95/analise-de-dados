```python
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds
```

```python
corte = diamonds[['cut', 'price']].groupby('cut').agg('mean').reset_index()
ordenado = corte.sort_values(by='price',ascending=True)


```

```python
grafico = sns.barplot(data=ordenado, x='cut', y='price', order=ordenado['cut'], palette='pastel')
grafico.set(title='Preço Médio do Diamante por tipo de corte', xlabel='Corte', ylabel='Preço Diamente (USD)')
```

2. Número de passageiros em dezembro por ano

```python
import seaborn as sns

data = sns.load_dataset("flights")
dezembro = data.query('month == "Dec"')
dezembro
```

```python
with sns.axes_style('darkgrid'):
    grafico = sns.lineplot(data=dezembro, x='year', y='passengers') #mudar parar linha
    grafico.set(title='Numero de Passageiros em Dezembro por ano', xlabel='Ano', ylabel='Passageiros')
```

3. Numero de passageiros por mês entre 1949 e 1959

```python
import seaborn as sns

data = sns.load_dataset("flights")
passageiros = data.query('year <= 1959')
```

```python
grafico = sns.lineplot(data=passageiros, x='month',y='passengers', hue='year', palette='pastel')
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
#grafico = sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='day', col='time', row='sex')
```

```python
grafico = sns.relplot(data=tips, x='total_bill', y='tip', hue='day', palette='dark')
```

```python
grafico.ax.set_title('Gorjetas', fontsize=12, fontweight='bold') # titulo do grafico
grafico.set_xlabels('Valor da conta (USD)', fontsize=10) #titulo para o eixo x
grafico.set_ylabels('valor da gorjeta (USD)', fontsize=10) # titulo para o eixo y
```

```python
grafico.fig #mostra a figura
```

```python
grafico.legend.set_title('Dias da Semana', prop={'size':10, 'weight':'bold'}) #mudar o titulo da legenda
grafico.fig
```

```python
# pra definir a altura e a largura da imagem é feita em polegada (10cm / 2.54)
altura = 10 / 2.54
largura = 40 / 2.54 
grafico.fig.set_size_inches(w=largura, h=altura)
grafico.fig
```

```python
grafico.fig.savefig(fname='gorjetas.png', bbox_inches='tight') # salva o arquivo em png
grafico.fig.savefig(fname='gorjetas.pdf', bbox_inches='tight')
```
