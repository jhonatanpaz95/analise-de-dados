```python
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds
```

```python
corte = diamonds[['cut', 'price']].groupby('cut').agg('mean').reset_index()
ordenado = corte.sort_values(by='price',ascending=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")
grafico = sns.barplot(data=ordenado, x='cut', y='price', order=ordenado['cut'], palette='pastel', )
grafico.set_title('Preço Médio do Diamante por tipo de corte', fontsize= 12, fontweight='bold' )
grafico.set(xlabel= 'Tipo do Corte', ylabel='Preço (USD)' )
```

```python
#Preços outliers de diamante

grafico = sns.boxplot(x=diamonds['price'], y=diamonds['cut'])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

with sns.axes_style('whitegrid'):
  grafico = sns.boxplot(data=diamonds, x="price",)
  grafico.set_title("Distribuição de Preços dos Diamantes")
  grafico.set(xlabel='Preço (USD)')
```

Correlação entre o do preço do diamante com seu peso (carat) agrupados por sua transparêcia (clarity)

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

with sns.axes_style('whitegrid'):
  grafico = sns.scatterplot(data=diamonds, x='carat', y='price', hue='clarity', palette='viridis', alpha=0.6)
  grafico.set_title('Correlação entre Preço e Peso (Carat) dos Diamantes Agrupados por Claridade (Clarity)')
  grafico.set(xlabel='Peso (Carat)', ylabel='preço(USD)')
  grafico.legend(title='Clarity')
  
```

2. Número de passageiros em dezembro por ano

```python
import seaborn as sns

data = sns.load_dataset("flights")
dezembro = data.query('month == "Dec"')
dezembro
```

```python
with sns.axes_style('darkgrid'):
    grafico = sns.lineplot(data=dezembro, x='year', y='passengers')
    grafico.set(title='Numero de Passageiros em Dezembro por ano', xlabel='Ano', ylabel='Passageiros')
```

```python
# grafico de area

import matplotlib.pyplot as plt

with sns.axes_style('whitegrid'):
    grafico = sns.FacetGrid(data=dezembro, palette='pastel')  # Corrigido aqui
    grafico.map(sns.lineplot, 'year', 'passengers')
    grafico.map(plt.fill_between, 'year', 'passengers', alpha=0.5)
    grafico.set(title= 'Numero de Passageiros em Dezembro por Ano',xlabel='Ano', ylabel='Numero de Passageiros')
    grafico.fig.set_size_inches(w=20/2.54, h=15/2.54)
```

3. Numero de passageiros por mês entre 1949 e 1959

```python
import seaborn as sns

data = sns.load_dataset("flights")
passageiros = data.query('year <= 1959')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
grafico = sns.lineplot(data=passageiros, x='month',y='passengers', hue='year', palette='bright')
grafico.set_title('Numero de Passageiros por Mês entre 1949 e 1959', fontsize= 12, fontweight='bold')
grafico.set(xlabel='Mês', ylabel='Numero de Passageiros por Ano')

# Ajustando o título da legenda
legenda = grafico.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
legenda.set_title('Anos', prop={'size': 10, 'weight': 'bold'})
```
